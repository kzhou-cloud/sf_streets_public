```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
# Load language clusters from file and apply clustering
import json

```

```python
combined_df = pd.read_csv("combined_df_gpt_4.1_classified.tsv", sep="\t")
combined_df.groupby("city").count()
combined_df['language_gpt'] = combined_df['language_gpt'].fillna("ERROR")
combined_df['language_gpt'] = combined_df.apply(lambda x: "is_numbered" if x['is_numbered'] else x['language_gpt'], axis=1)
combined_df[combined_df['language_gpt']=='IS_NUMBERED']
combined_df.groupby("language_gpt").count().sort_values("street_name", ascending=False).head(20)
```

```python
# Load cluster definitions from JSON file
with open('data/language_clusters.json', 'r') as f:
    cluster_data = json.load(f)

# Build reverse mapping: language -> cluster
LANGUAGE_CLUSTERS = {}
for cluster, languages in cluster_data.items():
    for lang in languages:
        LANGUAGE_CLUSTERS[lang] = cluster

print(f"\nLoaded {len(LANGUAGE_CLUSTERS)} language mappings into {len(cluster_data)} clusters")

# Apply clustering (unmapped values go to 'Other')
combined_df['language_cluster'] = combined_df['language_gpt'].map(
    lambda x: LANGUAGE_CLUSTERS.get(x)
)
```

```python
(combined_df.groupby("language_cluster").count().sort_values("street_name", ascending=False)/len(combined_df)*100).round()
```
